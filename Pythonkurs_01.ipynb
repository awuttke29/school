## 1. Getting started
Der vermutlich bekannteste Start zum Lernen einer Programmiersprache ist 'Hello World'. Dabei wird die Ausgabefunktion, bei Python *print*, genutzt um "Hello World!" ausgeben zu lassen:

Eine Codezeile führt ihr aus, indem ihr diese markiert und "Shift + Enter" drückt.

```python
print('Hello World!')
```

Das geht allerdings nicht nur mit Textpassagen, auch Strings genannt, sondern beispielsweise auch mit dem Wert von sogenannten Variablen. Wir definieren hier eine Variable u als 42 und lassen uns daraufhin den Wert von u ausgeben:

```python
u=42
print(u)
```

Wir haben schon die `print()`-Funktion kennengelernt. Man kann allerdings auch selbst Funktionen definieren oder aus Pythonpackages (Bibliotheken) laden. Zum Beispiel kann man das Produkt zweier Zahlen mit folgender selbst definierter Funktion berechnen:

```python
def produkt(zahl1,zahl2):
    ergebnis = zahl1 * zahl2
    return ergebnis
```

`def` ist ein sogenanntes Schlüsselwort. Schlüsselwörter sind in der Programmiersprache festvergebene Wörter für die Programmiersprache. Nach einem `def` erwartet Python einen Namen für eine eigene Methode. In diesem Fall "produkt". Nach der Bennenung der Methode folgt eine offene und geschlossene Klammer, in diese können die notwendigen Eingabewerte eingetragene werden. Mithilfe der Variablen "zahl1" und "zahl2" kann das "ergebnis" berechnet werden. Zuletzt erfolgt die Anweisung `return` welche das berechnete "ergebnis" zurückgibt. 

Jetzt wenden wir diese Funktion auf die Zahl 5 und 3 an und lassen uns das Ergebnis mit dem `print()`-Befehl ausgeben:

```python
print(produkt(5,3))
```

## 2. Bedingte Anweisungen

In einer Programmiersprache versteht man unter einer bedingten Anweisung, wie man Verzweigungen auch nennt, Codeteile, die unter einer Bedingung ausgeführt werden. Liegt die Bedingung nicht vor, wird dieser Code nicht ausgeführt. Anders ausgedrückt: Eine Verzweigung legt fest, welcher von zwei (oder auch mehr) Programmteilen (Alternativen) in Abhängigkeit von einer (oder mehreren) Bedingungen ausgeführt wird. Bedingte Anweisungen und Verzweigungen werden in Programmiersprachen (ebenso wie die Schleifen) den Kontrollstrukturen zugerechnet, weil mit ihrer Hilfe ein Programm auf verschiedene Zustände, die sich aus Eingaben und Berechnungen ergeben, reagieren kann.

**Ein erstes Beispiel**

Die folgende Methode untersucht, ob es sich um eine positive, negative oder neutrale Zahl (0) handelt

```python
def vorzeichen(zahl):
    if zahl>0:                  #Bedingung
        return "positiv +"      #bedingte Anweisung1
    elif zahl<0:                #weitere Bedingung: elif ist die Abkürzung für "else if"
        return "negativ -"      #bedingte Anweisung2
    else:                       #anderenfalls
        return "neutral 0"      #bedingte Anweisung3
```

```python
print(vorzeichen(0))
```
